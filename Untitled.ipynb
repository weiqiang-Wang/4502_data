# Plotting Functions

```python
def multi_line_plot(all_data, top_labels, column="Confirmed"):
    f = plt.figure()
    
    ax = plt.axes()
    # Tell matplotlib to interpret the x-axis values as dates
    ax.xaxis_date()
    ax.set_title(column + " by State")
    
    f.set_figwidth(20)
    f.set_figheight(10)
    # Make space for and rotate the x-axis tick labels
    f.autofmt_xdate()
            
    for state in all_data['state'].unique():
        this_data = all_data[all_data['state'] == state]
        
        date_values = this_data['date']
        confirmed_values = this_data[column]
        
        plt.plot(date_values, confirmed_values, linewidth=2, label=state if state in top_labels else "_nolegend_")

    ax.legend(loc='upper left', title="State")
    plt.show()
```

```python
def plot_clusters(all_data, centroids, kmeans, figsize=(20, 10)):
    plt.figure(figsize = figsize)
                      
    plt.scatter(all_data.iloc[:, 0], all_data.iloc[:, 1], c= kmeans.labels_.astype(float), s=50, alpha=0.5)
    plt.scatter(centroids[:, 0], centroids[:, 1], c='red', s=50)
    plt.show()
```
