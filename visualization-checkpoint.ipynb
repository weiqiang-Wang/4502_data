```python
### Helper functions
def shape_file_to_df(filepath):
    shape = shp.Reader(filepath)
    
    fields = [x[0] for x in shape.fields][1:]
    records = [list(i) for i in shape.records()]
    shps = [s.points for s in shape.shapes()]
    
    df = pd.DataFrame(columns=fields, data=records)
    df = df.assign(coords=shps)
    
    return df
```

```python
# Calculate clusters from previous data

us_confirmed_and_deaths_df = pd.read_csv(filepath + "country_data/us_simplified_csv.csv")
# Convert str dates to datetime objects
us_confirmed_and_deaths_df.Date = pd.to_datetime(countries_aggregated_data.Date, format="%Y-%m-%d")
# print(us_confirmed_and_deaths_df.columns.values)

# Get most recent data
sub_df = us_confirmed_and_deaths_df[['Admin2', 'Province/State', 'Confirmed', 'Deaths']]
# print(sub_df)
most_recent_us_confirmed_and_deaths_df = sub_df.groupby(['Admin2', 'Province/State'], as_index=False)
most_recent_us_confirmed_and_deaths_df = most_recent_us_confirmed_and_deaths_df.agg({'Admin2': 'last', 'Province/State': 'last', 'Confirmed': 'sum', 'Deaths': 'sum'})
most_recent_us_confirmed_and_deaths_df = most_recent_us_confirmed_and_deaths_df.groupby('Province/State', as_index=False).agg({'Confirmed': 'sum', 'Deaths': 'sum'})

print(most_recent_us_confirmed_and_deaths_df)

# Cluster based on confirmed and death count
us_confirmed_and_deaths_kmeans = KMeans(n_clusters=3).fit(most_recent_us_confirmed_and_deaths_df[['Confirmed', 'Deaths']])
centroids = us_confirmed_and_deaths_kmeans.cluster_centers_
print(centroids)
```

# Plotting functions

```python
def line_plot_by_country(country_data, top_labels, column="Confirmed"):
    f = plt.figure()
    
    ax = plt.axes()
    # Tell matplotlib to interpret the x-axis values as dates
    ax.xaxis_date()
    ax.set_title(column + " by Country")
    
    f.set_figwidth(20)
    f.set_figheight(10)
    # Make space for and rotate the x-axis tick labels
    f.autofmt_xdate()
            
    for country in country_data.keys():
        this_country_data = country_data[country]
        
        date_values = this_country_data.Date
        confirmed_values = this_country_data[column]
        
        plt.plot(date_values, confirmed_values, linewidth=2, label=country if country in top_labels else "_nolegend_")

    ax.legend(loc='upper left', title="Country")
    plt.show()
```

```python
def plot_map(sf, x_lim = None, y_lim = None, figsize = (20,10)):
    plt.figure(figsize = figsize)
    id=0
    for shape in sf.shapeRecords():
        x = [i[0] for i in shape.shape.points[:]]
        y = [i[1] for i in shape.shape.points[:]]
        plt.plot(x, y, 'k')
        
        if (x_lim == None) & (y_lim == None):
            x0 = np.mean(x)
            y0 = np.mean(y)
            plt.text(x0, y0, id, fontsize=10)
        id = id+1
    
    if (x_lim != None) & (y_lim != None):     
        plt.xlim(x_lim)
        plt.ylim(y_lim)
```

```python
def plot_clusters(all_data, centroids, kmeans, figsize=(20, 10)):
    print(all_data)
    f = plt.figure(figsize = figsize)
        
    ax = plt.axes()
                  
    plt.scatter(all_data.iloc[:, 0], all_data.iloc[:, 1], label=all_data.groupby('Province/State').agg({'cluster': 'last'}), c= kmeans.labels_.astype(float), s=50, alpha=0.5)
    plt.scatter(centroids[:, 0], centroids[:, 1], c='red', s=50)
    
    for i, label in enumerate(all_data['cluster']):
#         print(i, label)
        plt.text(all_data.iloc[:, 0][i], all_data.iloc[:, 1][i], label)
    
    ax.legend(loc='upper left', title="State")
    plt.show()
```

# Calling Plotting

```python
most_recent_us_confirmed_and_deaths_df['cluster'] = us_confirmed_and_deaths_kmeans.labels_

cats = ["Low", "High", "Medium"]
# most_recent_us_confirmed_and_deaths_df = most_recent_us_confirmed_and_deaths_df.sort_values(by='cluster')
most_recent_us_confirmed_and_deaths_df['cluster'] = most_recent_us_confirmed_and_deaths_df['cluster'].apply(lambda x: cats[x])

plot_clusters(most_recent_us_confirmed_and_deaths_df[['Confirmed', 'Deaths', 'Province/State', 'cluster']], centroids, us_confirmed_and_deaths_kmeans)

# most_recent_us_confirmed_and_deaths_df.groupby('Province/State', as_index=True).agg({'Province/State': 'last', 'cluster': 'last'}).iloc[:, 1]
```
